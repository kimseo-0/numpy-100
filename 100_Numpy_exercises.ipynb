#### 38. Consider a generator function that generates 10 integers and use it to build an array (‚òÖ‚òÜ‚òÜ)
> 10Í∞úÏùò Ï†ïÏàòÎ•º ÏÉùÏÑ±ÌïòÎäî Ï†úÎÑàÎ†àÏù¥ÌÑ∞ Ìï®ÏàòÎ°ú Î∞∞Ïó¥ÏùÑ ÎßåÎìúÏÑ∏Ïöî.

````python
# iterable
def generate():
    for x in random.randint(0, 10, 10):
        yield x

Z = np.fromiter(generate(),dtype=int,count=-1)
print(Z)
````
