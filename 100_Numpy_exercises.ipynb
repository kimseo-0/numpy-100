#### 16. How to add a border (filled with 0's) around an existing array? (★☆☆)

````python
Z = np.ones((5, 5))

# pad_width 만큼 숫자 0 으로 채움
Z = np.pad(Z, pad_width=1, mode='constant', constant_values=0)
print(Z)
````

````python
# edge
# pad_width 만큼 엣지가 확장
Z = np.ones((5, 5))
Z = np.pad(Z, pad_width=0, mode='edge') 
print(Z)
````

````python
# maximum
# 해당 행 또는 열의 maximum 값을 엣지에 작성
# stat_length = 해당 좌표를 기준으로 해당 범위 만큼에서 maximum 값을 찾는다

Z = np.arange(0,9).reshape((3,3))
Z = np.pad(Z, pad_width=1, mode='maximum', stat_length = 2)
print(Z)
````

````python
# minimum
Z = np.arange(0,9).reshape((3,3))

# 해당 행 또는 열의 minimum 값을 엣지에 작성
# stat_length = 해당 좌표를 기준으로 해당 범위 만큼에서 minimum 값을 찾는다

Z = np.arange(0,9).reshape((3,3))
Z = np.pad(Z, pad_width=1, mode='minimum', stat_length = 1)
print(Z)

# mean : 평균
# median : median 평균
````

````python
Z = np.arange(0,9).reshape((3,3))

# reflect
# 엣지가 기존 벡터를 반사된 형태
Z = np.pad(Z, pad_width=2, mode='reflect', reflect_type = 'odd')
print(Z)
````

````python
Z = np.arange(0,9).reshape((3,3))

# symmetric
# 엣지가 기존 벡터를 반사된 형태
Z = np.pad(Z, pad_width=1, mode='symmetric', reflect_type = 'even')
print(Z)
````

````python
Z = np.arange(0,9).reshape((3,3))

# wrap
# 행과 열의 반대편 끝에 있는 원소 값으로 빈 곳 채우기
Z = np.pad(Z, pad_width=2, mode='wrap')
print(Z)
````

````python
# empty

Z = np.arange(0,9).reshape((3,3))
Z = np.pad(Z, pad_width=1, mode='empty')
print(Z)
````

#### 17. What is the result of the following expression? (★☆☆)
```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
print("0 * np.nan =>", 0 * np.nan)
print("np.nan == np.nan =>", np.nan == np.nan)

 # nan 은 숫자가 아니므로 비교 자체가 성립하지 않음
print("np.inf > np.nan", np.inf > np.nan)
print("np.inf < np.nan", np.inf < np.nan)

print("np.nan - np.nan", np.nan - np.nan)
print("np.nan in set([np.nan])", np.nan in set([np.nan]))
print("0.3 == 3 * 0.1", 0.3 == 3 * 0.1)
````

#### 18. Create a 5x5 matrix with values 1,2,3,4 just below the diagonal (★☆☆)

diagonal : 대각 행렬




````python
# np.arange(4) => [0, 1, 2, 3, 4] + 1 > 각 원소에 +1
# np.arrange(1, 5) 로 대체 가능
# k 는 padding 을 만드는데, k > 0 이면 아래 행, k < 0 이면 위에 행에 생긴다
v = np.diag(np.arange(4) + 1, k = 1) # k = 1 도 가능
print(v)
````

#### 22. Normalize a 5x5 random matrix (★☆☆)

````python
Z = np.random.random((5, 5))
Z = (Z - np.mean (Z)) / (np.std (Z))
print(Z)
````

````python
Z = np.random.random((5, 5))
Z = (Z - np.min(Z)) / (np.max(Z) - np.min(Z))
print(Z)
````

#### 23. Create a custom dtype that describes a color as four unsigned bytes (RGBA) (★☆☆)

````python
color = np.dtype([("r", np.ubyte), ("g", np.ubyte), ("b", np.ubyte), ("a", np.ubyte)])

# 배열 선언
pixels = np.array([
    (255, 0, 0, 255),   # 빨강
    (0, 255, 0, 255),   # 초록
    (0, 0, 255, 128)    # 반투명 파랑
], dtype=color)

print(pixels[0]["r"])  # 255
print(pixels[2]["a"])  # 128
````
